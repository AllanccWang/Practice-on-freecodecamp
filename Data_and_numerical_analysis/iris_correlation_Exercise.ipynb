```python
# Exercise2 from https://milliams.com/courses/applied_data_analysis/Correlation.html
from pandas import DataFrame
from sklearn.datasets import load_iris

iris, iris_target = load_iris(as_frame=True, return_X_y=True)
iris.head()
```

```python
corr = iris.corr()
```

```python
%matplotlib inline

import seaborn as sns

sns.heatmap(corr, vmin=-1.0, vmax=1.0, cmap="RdBu")
```

```python
#If you see a strong diagonal line it means that those variables are correlated in this data set.
#It it's more of a blob or a flat horizontal or vertical line then that suggests a low correlation.
from pandas.plotting import scatter_matrix

a = scatter_matrix(iris, figsize=(16, 16), c=iris_target)
```
