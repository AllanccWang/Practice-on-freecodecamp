```python
#problem from https://milliams.com/courses/data_analysis_python/Final%20exercise.html
import pandas as pd
import seaborn as sns

sns.set_theme()
titanic = pd.read_csv("https://milliams.com/courses/data_analysis_python/titanic.csv")
```

```python
titanic
```

```python
#Find the average age of all people on board
average_age_of_people = titanic["age"].mean()
average_age_of_people
```

```python
#Use a filter to select only the males
filter_to_male = titanic[titanic["gender"]=="male"]
filter_to_male
```

```python
#Find the average age of the males on board
average_age_of_male = filter_to_male["age"].mean()
average_age_of_male
```

```python
#Use a filter to select only the people in 3rd class
filter_to_3Class = titanic[titanic["class"]=="3rd"]
filter_to_3Class
```

```python
#Create a DataFrame which only contains the passengers on the ship (those in first, second or third class)
filter_to_on_Ship = titanic[(titanic["class"]=="3rd") | (titanic["class"]=="2nd") | (titanic["class"]=="1st")]
filter_to_on_Ship
```

```python
#Plot and compare the distribution of ages for males and females.
sns.displot(
    data=titanic,
    x="age",
    hue="gender",  # Added this line
    stat="density",  # Added this line
    common_norm=False,  # Added this line
)
```

```python
#And how does this differ by class?
sns.relplot(
    data=titanic,
    x="age",
    y="class",
    hue="gender",
).set(
    xlabel="age",
    ylabel="class",
)
```

```python
#Calculate the percentage that survived within each class
survived = titanic[titanic["survived"]=="yes"]
survived
```

```python
titanic["class"].drop_duplicates()
```

```python
#Calculate the percentage that survived in class, 3rd
percentage_survived_3rd = len(titanic[(titanic["class"]=="3rd") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("3rd class :",percentage_survived_3rd)
#Calculate the percentage that survived in class, 2nd
percentage_survived_2nd = len(titanic[(titanic["class"]=="2nd") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("2nd class :",percentage_survived_2nd)
#Calculate the percentage that survived in class, 1st
percentage_survived_1st = len(titanic[(titanic["class"]=="1st") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("1st class :",percentage_survived_1st)
#Calculate the percentage that survived in class, engineering crew
percentage_survived_engineering_crew = len(titanic[(titanic["class"]=="engineering crew") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("engineering crew class :",percentage_survived_engineering_crew)
#Calculate the percentage that survived in class, victualling crew
percentage_survived_victualling_crew = len(titanic[(titanic["class"]=="victualling crew") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("victualling crew class :",percentage_survived_victualling_crew)
#Calculate the percentage that survived in class, restaurant staff
percentage_survived_restaurant_staff = len(titanic[(titanic["class"]=="restaurant staff") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("restaurant staff class :",percentage_survived_restaurant_staff)
#Calculate the percentage that survived in class, deck crew
percentage_survived_deck_crew = len(titanic[(titanic["class"]=="deck crew") & (titanic["survived"]=="yes")]) / len(survived) * 100.0
print("deck crew class :",percentage_survived_deck_crew)

print("total percentage :",percentage_survived_3rd+percentage_survived_2nd+percentage_survived_1st+percentage_survived_engineering_crew+
      percentage_survived_victualling_crew+percentage_survived_restaurant_staff+percentage_survived_deck_crew)
```

```python
#what the main factors predicting survival were
sns.relplot(
    data=titanic,
    x="age",
    y="class",
    hue="survived",
    style="gender",
).set(
    xlabel="class",
    ylabel="age",
)
#from below graph, the deck crew and the 1st,2nd class and age<40 are high survived rate
```
